```python
from math import sqrt, pi, cos
from tkinter import Tk, Canvas, Frame, BOTH
import random

class Case(Frame):
    
    #инициализируем входные данные
    def __init__(self, xStart, yStart, xEnd, yEnd, xCenter, yCenter, R):
        
        self.xStart = xStart
        self.yStart = yStart
        self.xEnd = xEnd
        self.yEnd = yEnd
        
        self.xCenter = xCenter
        self.yCenter = yCenter
        self.R = R
        
        super().__init__()
        
    #функция отрисовки прямых, окружности и точек пересечения
    def initUI(self):
        self.pack(fill=BOTH, expand=1)
 
        canvas = Canvas(self)
        
        #отрисовываем прямые
        for i in range(len(self.xEnd)):
            canvas.create_line(self.xStart, self.yStart, self.xEnd[i], self.yEnd[i])
            
        #отрисовываем окружность
        canvas.create_oval(self.xCenter-self.R, self.yCenter-self.R, self.xCenter+self.R, self.yCenter+self.R, outline='red')
        
        #отрисовываем точки пересечения
        for i in range(len(self.xCoordPer)):
            canvas.create_oval(self.xCoordPer[i]-2, self.yCoordPer[i]-2, self.xCoordPer[i]+2, self.yCoordPer[i]+2, outline='orange')
        canvas.pack(fill=BOTH, expand=1)
    
    #функция нахождения пересечения точек
    def Peresechenie(self):
        
        #списки для точек пересечения
        self.xCoordPer = []
        self.yCoordPer = []
        
        #запускаем цикл для нахождения пересечений окружности с прямыми
        for i in range(len(self.xEnd)):
            
            """Здесь используя два исходных уравнения:
            
                Ax + By + C = 0 - общее уравнение прямой
                (x - x0)^2 + (y - y0)^2 = R^2 - общее уравнение окружности, где x0,y0 - центр окружности
                
                выводим квадратное уравнение уравнение, выражая y из уравнения прямой и подставляем в уравнение
                окружности.
            """
            A = self.yStart - self.yEnd[i]
            B = self.xEnd[i] - self.xStart
            C = self.xStart*self.yEnd[i] - self.xEnd[i]*self.yStart

            result = ""
            
            # a,b,c - коэффициенты при x в квадратном уравнении
            a = 1 + (A**2/B**2)
            b = -2*self.xCenter + ((2*A*C)/(B**2)) + ((2*A*self.yCenter)/B)
            c = self.xCenter**2 + (C**2/B**2) + ((2*C*self.yCenter)/B) + self.yCenter**2 - self.R**2
            
            # считаем дискриминант
            discriminant = b*b - 4*a*c

            if discriminant > 0:
                x1 = (-b + sqrt(discriminant))/(2*a)
                x2 = (-b - sqrt(discriminant))/(2*a)

                y1 = (-A*x1 - C) / B
                y2 = (-A*x2 - C) / B
                result = "Есть две точки пересечения-"
                print(result,"(",x1, ",", y1,")","(",x2, ",", y2,")")
            elif (discriminant == 0):
                x1 = -b/(2*a)
                y1 = (-A*x1 - C) / B
                result = "Есть одна точка пересечения - "
                print(result,"(",x1, ",", y1,")")
            elif (discriminant < 0):
                result = "Точек пересечения нет"
                x1 = 0
                y1 = 0
                y2 = 0
                x2 = 0
            self.xCoordPer.append(x2)
            self.yCoordPer.append(y2)

# функция, которая генерирует точки Y для прямых от 0 до 30 градусов
def generationPoint(endX, endY):
    AC = endX
    AB = int(AC/cos(pi/6))
    
    coordsY = [i for i in range(endY, AB, 5)]
    coordsX = [500 for i in range(len(coordsY))]
    
    return coordsX, coordsY
        
def main():
    root = Tk()
    xCoord, yCoord = generationPoint(250, 0)
    case = Case(0,0,xCoord,yCoord,200,80,60)
    case.Peresechenie()
    case.initUI()
    
    root.geometry("500x500")
    root.mainloop()
        
if __name__ == '__main__':
    main()
```
